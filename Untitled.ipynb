```python
import pickle
def load_obj(name ):
    with open( name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
name2id = load_obj('name2id')
A_pred = load_obj('params')
```

```python



res = []
with open('./test-public.csv', 'r') as f:
    next(f)
    for line in f:
        _, h, t = line.split(',')
        h = int(h)
        t = int(t)
        
        res.append(A_pred[name2id[h]][name2id[t]].detach().numpy())
        #print(int(h), int(t))
```

```python
print(res[:3])
```

```python
out = 'your_name.csv'
with open(out, 'w') as o:
    o.write('Id,Predicted\n')
    for i, s in enumerate(res):
        o.write(f'{i+1},{s}\n')
print('Write output to', out)
```
